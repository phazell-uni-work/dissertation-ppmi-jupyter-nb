```python
import math
import numpy as np
import pandas
from scipy.stats import ttest_ind

original_data = pandas.read_csv('../working_data/updrsii_all_q_p_s.csv')

original_data.head()

# ditch prev_sigfall

fallers = original_data.query('SIGFALL == 1.0')
fallers = fallers.drop(['PREV_SIGFALL','SIGFALL'], axis=1)
non_fallers = original_data.query('SIGFALL == 0.0')
non_fallers = non_fallers.drop(['PREV_SIGFALL','SIGFALL'], axis=1)
```

```python
non_fallers.head()
```

```python
qs = non_fallers.columns

significant_qs = []

for q in qs:
    p = ttest_ind(fallers[q], non_fallers[q], equal_var=False)
    if p.pvalue < 0.05:
        print(q)
        print(p)
        significant_qs.append(q)
```

```python
for q in qs:
    p = ttest_ind(fallers[q], non_fallers[q], equal_var=False)
    print(f'{q},{p.statistic},{p.pvalue}')
```

```python
# Q's & SIGFALLS
a_qs = significant_qs + ['SIGFALL']
a_data_set = original_data[a_qs]
a_data_set.to_csv('../working_data/updrsii_select_s.csv', index=False)

b_qs = significant_qs + ['PREV_SIGFALL','SIGFALL']
b_data_set = original_data[b_qs]
b_data_set.to_csv('../working_data/updrsii_select_p_s.csv', index=False)
```
