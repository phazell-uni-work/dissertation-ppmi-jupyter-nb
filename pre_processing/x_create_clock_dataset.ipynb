```python
import pandas
import numpy

df_participant_data = pandas.read_csv('../source_data/participant_main.csv')
df_clock_raw = pandas.read_csv('../source_data/Clock_Drawing.csv')
df_age_at_visit = pandas.read_csv('../source_data/Age_at_visit.csv')

def age_at_visit(patno,event_id):
    patno = int(patno)
    try:
        return df_age_at_visit[(df_age_at_visit['PATNO'] == patno) & (df_age_at_visit['EVENT_ID'] == event_id)]['AGE_AT_VISIT'].values[0]
    except:
        print(f'{patno}:{event_id} not found')
        
        
```

```python
df_participant_data.head()
```

```python
# Add desired data to clock_data

def clock_data_add_age(row):
    p = row['PATNO']
    e = row['EVENT_ID']
    return age_at_visit(p,e)

def add_cohort(row,cohort):
    p = row['PATNO']
    try:
        x = df_participant_data[df_participant_data['PATNO'].isin([p])][cohort].values[0]
        return x
    except:
        print(f'PATNO {p} not found') # Will print every PATNO excluded in create_main_data.ipynb
    
df_clock_raw['AGE_AT_EVENT'] = df_clock_raw.apply (lambda row: clock_data_add_age(row), axis=1)
df_clock_raw['COHORT_PD'] = df_clock_raw.apply (lambda row: add_cohort(row,'COHORT_PD'), axis=1)
df_clock_raw['COHORT_PRODROMAL'] = df_clock_raw.apply (lambda row: add_cohort(row,'COHORT_PRODROMAL'), axis=1)
df_clock_raw['COHORT_HC'] = df_clock_raw.apply (lambda row: add_cohort(row,'COHORT_HC'), axis=1)
```

```python
df_clock_usable = df_clock_raw[df_clock_raw['COHORT_PD'].isin([0,1])]


df_clock_data = df_clock_usable[['PATNO', 'INFODT','AGE_AT_EVENT','CLCKPII',
       'CLCK2HND', 'CLCKNMRK', 'CLCKNUIN', 'CLCKALNU', 'CLCKNUSP', 'CLCKNUED',
       'CLCKTOT','COHORT_PD',
       'COHORT_PRODROMAL', 'COHORT_HC']]


df_clock_model_pd = df_clock_usable[['CLCKPII','CLCK2HND', 'CLCKNMRK', 'CLCKNUIN', 'CLCKALNU', 'CLCKNUSP', 'CLCKNUED','CLCKTOT','COHORT_PD']]
df_clock_model_hc = df_clock_usable[['CLCKPII','CLCK2HND', 'CLCKNMRK', 'CLCKNUIN', 'CLCKALNU', 'CLCKNUSP', 'CLCKNUED','CLCKTOT','COHORT_HC']]
```
