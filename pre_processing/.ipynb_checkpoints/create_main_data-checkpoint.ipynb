```python
import pandas
import numpy

df_participent_status = pandas.read_csv('../source_data/Participant_Status.csv')
```

```python
# print(participant_data.COHORT.value_counts())
```

```python
# Add new columns

def is_pd(row):
    cohort = row['COHORT']
    if int(cohort) == 1:
        return 1
    else:
        return 0
    
def is_prodromal(row):
    cohort = row['COHORT']
    if int(cohort) == 4:
        return 1
    else:
        return 0

def is_control(row):
    cohort = row['COHORT']
    if int(cohort) == 2:
        return 1
    else:
        return 0

df_participent_status['COHORT_PD'] = df_participent_status.apply(lambda row: is_pd(row), axis=1)
df_participent_status['COHORT_PRODROMAL'] = df_participent_status.apply(lambda row: is_prodromal(row), axis=1)
df_participent_status['COHORT_HC'] = df_participent_status.apply(lambda row: is_control(row), axis=1)
```

```python
df_participent_status.head()
```

```python
# filter

# 1 = ParkinsonsDisease, 2 = HealthyControl, 4 = Prodromal

df_participant_usable = df_participent_status[df_participent_status['ENROLL_STATUS'].isin(['enrolled','Complete', 'Enrolled','complete']) & df_participent_status['COHORT_DEFINITION'].isin(["Healthy Control", "Parkinson's Disease", "Prodromal"])]

df_participant_data = df_participant_usable[['PATNO','ENROLL_DATE','ENROLL_AGE','COHORT_PD', 'COHORT_PRODROMAL','COHORT_HC']]
df_participant_data.to_csv('../source_data/participant_main.csv')

```
